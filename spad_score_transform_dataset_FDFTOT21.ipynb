'''
Les informations sur la base de données nous indique qu'il y a 23 variables : la première variable indiquant les identifiants et les 22 variables suivantes représentant les caractéristiques du dataset (la dernière variable (IFI+) s'agissant de la variable cible) :
 - il n'y a pas de valeurs manquantes dans les 21 premières variables ;
 - la variable IFI+ contient deux modalités mais les valeurs manquantes sont aussi présentes.
 
Quand on affiche le dataset, les données n'ont pas de délimiteurs si on ne prend pas en compte la première colonne. Les valeurs de la deuxième colonne sont dans une chaîne de caractères. Pour bien lire les valeurs en python, il va falloir créer un programme pour séparer les valeurs deux par deux et les enregistrer dans un nouveau fichier texte.

'''

```python
# import du dataset
data_orig = pd.read_csv("/content/drive/MyDrive/spad_score_method_1/spad_projection_FDFTOT21/FDFTOT21.DAD",delim_whitespace="\t" , header = None, low_memory=False, dtype=str)
data = data_orig.copy()
data
```

```python
# les noms des variables

colonnes =(["IDENT", "TypeHabi", "GeoFaz_Scl", "GeoFaz_cl", "ImmoFaz_Scl", "ImmoFaz_cl", "Anciennete",
            "Logement_prix_m2", "Mediane_RFR", "Sexe", "Prenom_Phase_scl_H", "Prenom_Phase_scl_F",
            "Prenom_Phase_cl_H", "Prenom_Phase_cl_F", "Age_H", "Age_F", "Affinite_1", "Affinite_2",
            "Prenom_effectif", "Potentiel_prenom", "Onoma-Phase_Scl", "Onoma-Phase_CL", "Top_mon"])

```

```python
len(colonnes)
```

```python
# création d'un dataframe vide ayant la même dimension que l'objet data
df = pd.DataFrame(columns=colonnes, index = range(data.shape[0])) 
```

```python
# vérification des valeurs manquantes. A la suite du premier analyse de la base de données
# on se rend compte que les longueurs des chaînes de caractères sont 44 ou 45.
taille_42 =0
taille_43 =0
for i in data_orig.iloc[:,1]:
    if len(i)==42:
        taille_42 = taille_42 + 1
    if len(i)==43:
        taille_43= taille_43 + 1
print("nombre de ligne ayant une longueur égale à 42 : " ,taille_42)
print("nombre de ligne ayant une longueur égale à 43 : ",taille_43)
print("somme : ", taille_42+taille_43)
```

```python
# La deuxième analyse nous dévoile avec certitude les valeurs manquantes présentent à la
# dernière colonne. Donc, il va falloir créer un petit programme pour séparer deux par deux
# toutes les chaînes de caractères sans oublier les valeurs manquantes sur certaines lignes.
line = 0
for i in data.iloc[:,1]:
    for k in range(42):      
        if k%2 == 0:         
            df.iloc[line,1+int(k/2)]= i[k]+i[k+1]
    try:
        df.iloc[line,22] = i[42]
    except:
        print("valeur manquante")    
    #print(line)
    line= line + 1
```

```python
# stockage des identifiants (bien évidemment, pas besoin de les séparer)
df["IDENT"]=data.iloc[:,0]
```
